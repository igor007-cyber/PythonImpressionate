```python
import locale
import time 

locale.setlocale(locale.LC_TIME,'pt_BR.UTF-8')

tempo_atual_segundos = time.time()
print(f'tempo real: {tempo_atual_segundos} em segundos')

print('pausa')
time.sleep(5)
print('voltou')

tempo_agora = time.ctime(tempo_atual_segundos)
print(f'tempo local é: {tempo_agora}')

tempo_separado = time.localtime(tempo_atual_segundos)
print(f'tempo local é: {tempo_separado}')

print(time.strftime('%d %B %Y',tempo_separado))
print(time.strftime('%H %M %S',tempo_separado))
```

```python
import matplotlib.pyplot as plt

vendas = [1500,354,5000,9463,422]
meses = ['jan','fev','mac','maio','jun']

plt.plot(meses,vendas)
plt.ylabel('vendas')
plt.xlabel('meses')
plt.axis([0,5,0,max(vendas)+500])
plt.show()
```
